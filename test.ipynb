```python
import cv2, matplotlib
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# read an image
chair = cv2.imread('chair.jpg')
wallet = cv2.imread('wallet.jpg')
# show image format (basically a 3-d array of pixel color info, in BGR format)
print(chair,wallet)
```

```python
chair = cv2.cvtColor(chair, cv2.COLOR_BGR2RGB)
wallet = cv2.cvtColor(wallet, cv2.COLOR_BGR2RGB)
plt.imshow(chair)
```

```python
# convert image to grayscale
gray_chair = cv2.cvtColor(chair, cv2.COLOR_RGB2GRAY)
gray_wallet = cv2.cvtColor(wallet, cv2.COLOR_RGB2GRAY)
# grayscale image represented as a 2-d array
print(gray_wallet)
```

```python
plt.imshow(gray_wallet)
```

```python
average_color_per_row = np.average(chair, axis=0)
average_color = np.average(average_color_per_row, axis=0)
# convert back to uint8
average_color = np.uint8(average_color)
print(average_color)
```

```python
# create 100 x 100 pixel image with average color value
average_color_chair = np.array([[average_color]*100]*100, np.uint8)
 
plt.imshow(average_color_chair)
```

```python
_, threshold_chair = cv2.threshold(gray_chair, 90, 255, cv2.THRESH_BINARY)
 
# show image
threshold_chair = cv2.cvtColor(threshold_chair, cv2.COLOR_GRAY2RGB)
plt.imshow(threshold_chair)
```

```python
_, threshold_wallet = cv2.threshold(gray_wallet, 90, 255, cv2.THRESH_BINARY)
 
# show image
threshold_wallet = cv2.cvtColor(threshold_wallet, cv2.COLOR_GRAY2RGB)
plt.imshow(threshold_wallet)
```

```python
# 
threshold_diff = cv2.absdiff(threshold_wallet, threshold_chair)
plt.imshow(threshold_diff)
```

```python
chair_diff = cv2.absdiff(chair, wallet)
plt.imshow(chair_diff)
```

```python
# convert image to grayscale
gray_diff = cv2.cvtColor(chair_diff, cv2.COLOR_RGB2GRAY)
# grayscale image represented as a 2-d array
_, threshold_diff = cv2.threshold(gray_diff, 50, 255, cv2.THRESH_BINARY)
 
# show image
threshold_diff = cv2.cvtColor(threshold_diff, cv2.COLOR_GRAY2RGB)
plt.imshow(threshold_diff)
```
